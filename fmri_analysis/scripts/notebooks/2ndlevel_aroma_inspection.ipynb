```python
import nilearn
from nilearn import plotting
import nibabel as nb
from nilearn import image
from os import path
from glob import glob

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
new_2ndlevels = glob('/Users/henrymj/Documents/mounts/OAK/data/uh2/aim1/BIDS_scans/derivatives/2ndlevel/DPX/secondlevel-RT-True_beta-False_maps/*.nii.gz')
new_2ndlevels = [m for m in new_2ndlevels if 'tfile' not in m]
```

```python
glob('/Users/henrymj/Documents/mounts/OAK/data/uh2/aim1/BIDS_scans/derivatives/2ndlevel_7_28_20/DPX/secondlevel-RT-True_beta-False_maps/*.nii.gz')
```

```python
for file in new_2ndlevels:
    new_map = nb.load(file)
    hdr, aff = new_map.header, new_map.affine
    old_map = nb.load(file.replace('2ndlevel', '2ndlevel_7_28_20'))
    plotting.plot_img(new_map, cut_coords=(0,0,0), title=path.basename(file))
    plotting.plot_img(old_map, cut_coords=(0,0,0), title='old')
    diff_map = nb.Nifti1Image(new_map.get_fdata() - old_map.get_fdata(), aff, hdr)
    plotting.plot_img(diff_map, title='diff')
    plt.show()
    plt.close()
```
